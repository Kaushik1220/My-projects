```python
#in this project we are doing face detection using haar cascades 
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import math
!curl -o pic.png https://thingqbator.s3.ap-south-1.amazonaws.com/tqb/2018/08/pic.jpg
```

```python
img=cv2.imread('pic.png',1)
```

```python
cv2_imshow(img)
```

```python
face_cascade=cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img2
```

```python
faces=face_cascade.detectMultiScale(img2,1.3,5)
```

```python
for (x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
  roi_gray = img2[y:y+h, x:x+w]
  roi_color = img[y:y+h, x:x+w]
```
